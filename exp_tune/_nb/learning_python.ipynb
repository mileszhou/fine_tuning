```python
p= dict(zip(['a','b','c'],[1,2,3]))
print(p)
p['d']= 4
print(p)
for k,v in p.items(): print(k,v)
print(list(p.keys()))
print(list(p.values()))
q= dict(zip(list(p.keys()),[v**2 for v in p.values()]))
print(q)
i= iter(q)
print(i)
print(next(i))
for k,v in q.items(): print(k,v)    
```
